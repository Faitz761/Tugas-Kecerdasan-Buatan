```python
graph = {
    "bus": {
        "is-a": "kendaraan",
        "menggunakan": "bensin"
    },
    "mobil": {
        "is-a": "kendaraan",
        "menggunakan": "bensin"
    },
    "motor": {
        "is-a": "kendaraan",
        "menggunakan": "bensin"
    },
    "kendaraan": {
        "digunakan_untuk": "transportasi"
    },
    "transportasi": {
        "jenis": "darat"
    }
}

while True:
    tanya = input("\nTanya: ").lower()

    if tanya == "exit":
        break

    objek_ditemukan = None

    # cari objek dalam kalimat
    for objek in graph:
        if objek in tanya:
            objek_ditemukan = objek
            break

    if objek_ditemukan:
        relasi_ditemukan = False

        # cek apakah ada relasi yang disebut di pertanyaan
        for relasi in graph[objek_ditemukan]:
            if relasi in tanya:
                print("Jawaban:", graph[objek_ditemukan][relasi])
                relasi_ditemukan = True
                break

        # kalau relasi tidak ditemukan, tampilkan semua info objek
        if not relasi_ditemukan:
            print("Informasi tentang", objek_ditemukan)
            for r, v in graph[objek_ditemukan].items():
                print(r, ":", v)

    else:
        print("Objek tidak ditemukan.")
```
